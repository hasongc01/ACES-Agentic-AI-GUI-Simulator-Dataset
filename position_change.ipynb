```python
import pandas as pd
import numpy as np
```

```python
# set user path
path = '/Users/chohasong/Desktop/streamlit_datasets/position_change/'
```

```python
fitness_watch = pd.read_csv(path+'fitness_watch.csv')
```

```python
fitness_watch[['sku','assigned_position','original_price','original_rating','original_rating_count']]
```

```python
def random_position(df, seed):
    rng = np.random.default_rng(seed)   # change seed for a different shuffle
    n = len(df)

    new_pos = rng.permutation(n)       # 0..n-1 shuffled
    random_position = df.copy()
    random_position["assigned_position"] = new_pos
    random_position["position_in_experiment"] = new_pos  # keep same as assigned_position

    # optional: sort so the dataframe is in display order
    random_position = random_position.sort_values("assigned_position").reset_index(drop=True)

    return random_position
```

```python
fitness_watch_random_position_1 = random_position(fitness_watch,43)
```

```python
fitness_watch_random_position_1[['title','assigned_position','original_price','original_rating','original_rating_count']]
```

```python
# fitness_watch_random_position_1.to_csv(path+'fitness_watch_random_position_1.csv')
```

```python
fitness_watch_random_position_2 = random_position(fitness_watch,63)
```

```python
fitness_watch_random_position_2[['title','assigned_position','original_price','original_rating','original_rating_count']]
```

```python
fitness_watch_random_position_2.to_csv(path+'fitness_watch_random_position_2.csv')
```
